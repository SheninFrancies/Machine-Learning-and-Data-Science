Separate the Marriage column into MarriageStatus and RelationshipStatus:

````python
data_copy['MarriageStatus'] = data_copy['Marriage'].apply(lambda x: x.split('-')[0])
data_copy.head()
````

````python
categorical_variables = list(filter(lambda x: data_copy[x].dtype == 'O', data_copy.columns))
````

````python
data_copy['Marriage'].unique()
````

````python
relation = []
for data in data_copy['Marriage']:
  if len(data.split('-')) > 2:
    relation.append((data.split('-')[1:])[0])
  else:
    relation.append(np.NAN)

Relation = [str(x) for x in relation]
data_copy['PartnerStatus'] = Relation
data_copy['PartnerStatus'] = data_copy['PartnerStatus'].map({'civ': 'Civil', 'spouse': 'Absent', 'AF':'Military'})
data_copy['PartnerStatus'].replace(np.NaN, 'No Information', inplace=True)
data_copy.head()
````

````python
data_copy['PartnerStatus'].unique()
````

````python
data_copy['PartnerStatus'].isna().sum()
````

Drop the Marriage Column:

````python
data_copy.drop('Marriage', axis=1, inplace=True)
data_copy.head()
````

````python
data_copy['WorkClass'].unique()
````

````python
temp = []
for data in data_copy['WorkClass']:
  temp.append(data)

stripped = [data.strip() for data in temp]
data_copy['WorkClass'] = stripped
data_copy['WorkClass'].unique()
````

````python
# data_copy['WorkClass'].map({'Self-emp-not-inc':'Self Employed Uninc', 'State-gov':'State', 'Federal-gov':'Federal', 
#                             'Local-gov':'Local', '?':'No Information', 'Self-emp-inc': 'Self Employed Inc', 
#                             'Never-Worked':'Unemployed', 'Without-pay':'Unpaid Work'})
# data_copy['WorkClass'].unique()
````

````python
work = {'Self-emp-not-inc':'Self Employed', 'Private':'Private', 'State-gov':'Government', 'Federal-gov':'Government', 
        'Local-gov':'Government', '?':'No Information', 'Self-emp-inc': 'Self Employed', 
        'Without-pay':'Unpaid Work', 'Never-worked':'Unemployed'}

class MultipleValues(dict):
    def keylist(self, keys, value):
        for key in keys:
            self[key] = value

WorkClass = MultipleValues(work)
data_copy['WorkClass'] = data_copy['WorkClass'].map(WorkClass)
data_copy['WorkClass'].unique()
````

````python
data_copy.head()
````

Intermediate Save:

````python
data_worked = data_copy.copy()
data_worked.head(3)
````

````python
data_worked['Education'].unique()
````

It is fair to say that we can categorize this data into distinct sensible categories:
```r
1. Preschool, 9th, 7th-8th, 5th-6th, 10th, 1st-4th, 11th, 12th => School
2. HS-Grad => High School
3. Assoc-acdm, Assoc-voc => Associate
4. Bachelors, Some-college => Undergraduate
5. Masters => Graduate
6. Doctorate => Doctorate
7. Prof-School => Professional School
```

````python
stripped = [data.strip() for data in data_worked['Education']]
data_worked['Education'] = stripped
data_worked['Education'].unique()
````

````python
education = {'Bachelors':'Undergraduate', 'HS-grad':'High School', '11th':'School', 'Masters':'Graduate',
             '9th':'School', 'Some-college':'Undergraduate','Assoc-acdm':'Associate', 'Assoc-voc':'Associate', '7th-8th':'School', 
             'Doctorate':'Doctorate', 'Prof-school':'Professional School','5th-6th':'School', '10th':'School', '1st-4th':'School', 'Preschool':'School', '12th':'School'}

EducationClass = MultipleValues(education)

data_worked['Education'] = data_worked['Education'].map(EducationClass)
data_worked['Education'].unique()
````

````python
data_worked.head()
````

````python
EdNum = data_worked['EducationNumber'].unique()
np.sort(EdNum)
````

````python
data_worked['Occupation'].unique()
````

````python
stripped = [data.strip() for data in data_worked['Occupation']]
data_worked['Occupation'] = stripped
data_worked['Occupation'].unique()
````

````python
data_worked['Occupation'].replace('?', 'Unknown', inplace=True)
data_worked['Occupation'].unique()
````

We leave this column as it is. There is nothing much to gain by making any changes here.

````python
data_worked['Native Country'] = [data.strip() for data in data_worked['Native Country']]
data_worked['Native Country'].unique()
````

````python
to_replace = {'?':'Unknown', 'Hong':'Hong Kong', 'Trinadad&Tobago':'Trinidad and Tobago',
              'Holand-Netherlands':'Netherlands'}
for key, value in to_replace.items():
  data_worked['Native Country'] = data_worked['Native Country'].replace(key, value)

data_worked['Native Country'].unique()
````
